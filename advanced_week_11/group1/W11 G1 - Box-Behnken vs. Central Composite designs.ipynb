```python
import matplotlib.pyplot as plt
import numpy as np

def plot_numberRuns(n_c):
    # Parameters
    factors = np.arange(2, 8)  # 2 to 7 factors
    n_c = 3  # center points

    # Compute runs
    ccd_runs = [2**k + 2*k + n_c for k in factors]
    bbd_runs = [2*k*(k-1) + n_c for k in factors]

    # Plot
    x = np.arange(len(factors))
    width = 0.35

    plt.figure(figsize=(8, 5))
    plt.bar(x - width/2, ccd_runs, width, label='CCD', color='skyblue')
    plt.bar(x + width/2, bbd_runs, width, label='BBD', color='salmon')

    # Labels and title
    plt.xticks(x, factors)
    plt.xlabel('Number of Factors (k)')
    plt.ylabel('Number of Runs')
    plt.title('Comparison of Runs: CCD vs BBD (3 Center Points)')
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
```

## 1. Why DoE

* Gain as much information as possible
* As few experiments as posible



## Fractional and Full Factorial designs

* Frational is a subset of Full Factorial experiments
  * Reduces experiment amount
  * Reduces information
* Full factorial
  * More experiments
  * More information


## Handling curvature

* Two level full factorial designs cannot with curvature
* Increasing levels quicly increases amount of experiments
  * Third level with 3 factors: $3^3 = 24 $
  * Third level with 4 factors: $3^4 = 81 $

![](https://ars.els-cdn.com/content/image/3-s2.0-B9780128155639000124-f12-01-9780128155639.jpg)

## Response surface designs

* Specificly designed to handle curvature
* While keeping experiment count to a minimum

**Role of centerpoints?**
* Check for curvature
* Provide a measure of process stability and inherent variability







### Number of Runs
if $k$ is the number of factors
* CDD: $N_{runs} = 2^k + 2k + n_{center}$
* BBD: $N_{runs} = 2k(k-1) + n_{center}$

**Repeated centerpoints:**
* Typically 3-5 repetitions, including start, mid and end of testing

```python
plot_numberRuns(n_c=3)
```

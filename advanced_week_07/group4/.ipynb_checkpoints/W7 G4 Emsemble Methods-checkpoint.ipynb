## Working summary
* Decision trees are simple models used for classification, based on if/then qualifiers.
* Bagging algorithms uses multiple decision trees and decides based on majority voting.
* Boosting algorithms builds decision trees sequentially, weighing the influence of previous generations based on performance.

```python
from ucimlrepo import fetch_ucirepo 
from sklearn.ensemble import BaggingClassifier as Bagging
from sklearn.ensemble import AdaBoostClassifier as Boosting
from sklearn.ensemble import RandomForestClassifier as Random
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# fetch dataset
# Dataset - ID
# Breast Cancer - 17
# Iris - 53
# Abalone - 1
set = fetch_ucirepo(id=17) 
X = set.data.features.copy()
num_cols = X.select_dtypes(include=[np.number]).columns.tolist()
X_num = X[num_cols].to_numpy(dtype=float)
y = set.data.targets

X_train, X_test, y_train, y_test = train_test_split(X_num, y, test_size=0.75, random_state=42)
```

```python
# variable information 
#display(set.variables) 
```

```python
def LCfunc(func, X_train, X_test, y_train, y_test, label, target=0.99, max_iter=100):
    scores = []
    n_estimators = []
    score = 0
    n_est = 1
    while score < target and n_est <= max_iter:
        classifier = func(n_estimators = n_est, random_state=42).fit(X_train, y_train)
        score = classifier.score(X_test, y_test)
        scores.append(score)
        n_estimators.append(n_est)
        n_est += 1
    return pd.DataFrame({label: scores})
```

```python
max_est = 60
results = pd.DataFrame({"n_est": range(1,max_est,1)})
results = pd.concat([results,
                    LCfunc(Random, X_train, X_test, y_train, y_test, target=0.99, max_iter=max_est, label = "Random Forest"),
                    LCfunc(Bagging, X_train, X_test, y_train, y_test, target=0.99, max_iter=max_est, label = "Bagging"),
                    LCfunc(Boosting, X_train, X_test, y_train, y_test, target=0.99, max_iter=max_est, label = "Boosting")],
                    axis=1) 

results.plot(x='n_est', y=['Bagging', 'Random Forest', 'Boosting'], kind='line',
            xlabel = "Number of estimators (trees)", ylabel = "Accuracy", grid = True)
```
